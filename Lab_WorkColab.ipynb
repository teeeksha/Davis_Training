```
#Question 3: Remove Common Elements from Two Lists
#Problem Statement:
#Remove elements from the first list that are present in the second list.
list1 = [1, 2, 3, 4, 5, 6]
list2 = [2, 4, 6]

set2 = set(list2)
result = [x for x in list1 if x not in set2]

print(result)
```

```
#Question 4: Tuple Element Frequency
#Problem Statement:
#Count the frequency of each element in a tuple
t = (1, 2, 2, 3, 1, 4, 2)

freq = {}

for item in t:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

print(freq)
```

```
#Question 5: Convert List to Tuple and Remove Duplicates
#Problem Statement:Convert a list into a tuple after removing duplicate elements
lst = [1, 2, 2, 3, 4, 4, 5]

result = tuple(set(lst))

print(result)
```

```
#Question 6: Find Maximum and Minimum in Tuple
#Problem Statement:Find the maximum and minimum elements in a tuple without using built-in functions
t = (3, 7, 2, 9, 1, 5)

max_val = t[0]
min_val = t[0]

for num in t:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Maximum:", max_val)
print("Minimum:", min_val)
```

```
#Question 7: Set Difference and Intersection
#Problem Statement:Find the intersection and difference between two sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

intersection = A & B
difference = A - B

print("Intersection:", intersection)
print("Difference:", difference)
```

```
#Question 8: Check Subset
#Problem Statement:Check whether one set is a subset of another
A = {1, 2}
B = {1, 2, 3, 4}

if A.issubset(B):
    print("A is subset of B")
else:
    print("A is NOT subset of B")
```

```
#Question 9: Merge and Sort Unique Elements
#Problem Statement:Merge two lists and return a sorted list of unique elements.
list1 = [4, 2, 8, 1]
list2 = [2, 5, 1, 9]

result = sorted(set(list1 + list2))

print(result)
```

```
#Question 10: Tuple Pair Sum
#Problem Statement:Find all pairs in a tuple whose sum equals a given value.
t = (2, 4, 3, 5, 7, 8)
target = 7

pairs = []

for i in range(len(t)):
    for j in range(i + 1, len(t)):
        if t[i] + t[j] == target:
            pairs.append((t[i], t[j]))

print(pairs)
```

```
#Question 11: Remove Empty Sets from List
#Problem Statement:Remove all empty sets from a list of sets.
lst = [{1, 2}, set(), {3, 4}, set(), {5}]

result = [s for s in lst if s]

print(result)
```

```
#Question 12: Common Elements in Three Lists
#Problem Statement:Find common elements present in three lists.
l1 = [1, 2, 3, 4]
l2 = [2, 3, 5]
l3 = [2, 3, 6]

result = list(set(l1) & set(l2) & set(l3))

print(result)
```

```
#Question 13: Flatten Nested List
#Problem Statement:Flatten a nested list (1 level deep).
lst = [[1, 2], [3, 4], [5, 6]]

result = [item for sublist in lst for item in sublist]

print(result)

```

```
#Question 14: Symmetric Difference of Sets
#Problem Statement:Find elements that are in either of the sets but not in both.
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

result = A ^ B

print(result)
```

```
#Question 15: Tuple to Dictionary Mapping
#Problem Statement:Convert two tuples into a dictionary where one tuple contains keys and the other contains values.
keys = ('a', 'b', 'c')
values = (1, 2, 3)

result = dict(zip(keys, values))

print(result)
```
