```
#Question 3: Remove Common Elements from Two Lists
#Problem Statement:
#Remove elements from the first list that are present in the second list.
list1 = [1, 2, 3, 4, 5, 6]
list2 = [2, 4, 6]

set2 = set(list2)
result = [x for x in list1 if x not in set2]

print(result)
```

```
#Question 4: Tuple Element Frequency
#Problem Statement:
#Count the frequency of each element in a tuple
t = (1, 2, 2, 3, 1, 4, 2)

freq = {}

for item in t:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

print(freq)
```

```
#Question 5: Convert List to Tuple and Remove Duplicates
#Problem Statement:Convert a list into a tuple after removing duplicate elements
lst = [1, 2, 2, 3, 4, 4, 5]

result = tuple(set(lst))

print(result)
```
