```
#Find out greatest digit present in given number
#Input : 3723
#output : greatest digit -- 7
#Answer :
num_str = input("Enter a number: ")

greatest_digit = 0
for digit_char in num_str:
    digit = int(digit_char)
    if digit > greatest_digit:
        greatest_digit = digit

print(f"greatest digit -- {greatest_digit}")
for x in range (10):
      print (x)
```

```
#Question :- To find out factorial of a number
num = int(input("Enter a number: "))
fact = 1

if num < 0:
    print("Factorial does not exist")
elif num == 0:
    print("Factorial is 1")
else:
    for i in range(1, num + 1):
        fact = fact * i
    print("Factorial is:", fact)
```
