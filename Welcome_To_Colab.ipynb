```
#Find out greatest digit present in given number
#Input : 3723
#output : greatest digit -- 7
#Answer :
num_str = input("Enter a number: ")

greatest_digit = 0
for digit_char in num_str:
    digit = int(digit_char)
    if digit > greatest_digit:
        greatest_digit = digit

print(f"greatest digit -- {greatest_digit}")
for x in range (10):
      print (x)
```

```
#Question :- To find out factorial of a number
num = int(input("Enter a number: "))
fact = 1

if num < 0:
    print("Factorial does not exist")
elif num == 0:
    print("Factorial is 1")
else:
    for i in range(1, num + 1):
        fact = fact * i
    print("Factorial is:", fact)
```

```
#to find hcf of 2 numbers
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
if(num2 == 0):
  print("HCF :",num1)
else:
    while(num2 != 0):
      rem = num1 % num2
      num1 = num2
      num2 = rem
      print("HCF :",num1)
```

```
#WAP to display even factors of a number
num = int(input("Enter a number: "))

print("Even factors are:")

for i in range(1, num + 1):
    if num % i == 0 and i % 2 == 0:
        print(i)
```

```
#WAP to check a numbers of odd or even
def check_even_odd(num):
    if num % 2 == 0:
        return "Even number"
    else:
        return "Odd number"

# input from user
number = int(input("Enter a number: "))

# function call
result = check_even_odd(number)

# output
print(result)
```
